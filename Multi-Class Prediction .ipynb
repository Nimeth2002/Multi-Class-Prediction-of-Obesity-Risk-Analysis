**Data Preprocessing**

```python
for dataset in combine:
    dataset['Gender'] = dataset['Gender'].map({'Male': 0, 'Female': 1}).astype(int)
    dataset['family_history_with_overweight'] = dataset['family_history_with_overweight'].map({'no': 0, 'yes': 1}).astype(int)
    dataset['FAVC'] = dataset['FAVC'].map({'no': 0, 'yes': 1}).astype(int)
    dataset['CAEC'] = dataset['CAEC'].map({'no': 0, 'Frequently': 1, 'Sometimes': 2, 'Always': 3}).astype(int)
    dataset['SMOKE'] = dataset['SMOKE'].map({'no': 0, 'yes': 1}).astype(int)
    dataset['SCC'] = dataset['SCC'].map({'no': 0, 'yes': 1}).astype(int)
    dataset['CALC'] = dataset['CALC'].map({'no': 0, 'Frequently': 1, 'Sometimes': 2, 'Always': 3}).astype(int)
    dataset['MTRANS'] = dataset['MTRANS'].map({'Automobile': 0, 'Motorbike': 1, 'Public_Transportation': 2, 'Bike': 3, 'Walking': 4})
```

```python
combine[0].head()
```

```python
target = combine[0]["NObeyesdad"]
test_id = combine[1]["id"]

combine[0] = combine[0].drop(["id", "NObeyesdad"], axis=1)
combine[1] = combine[1].drop("id", axis=1)

dict_target = {'Insufficient_Weight':0,
                'Normal_Weight':1,
                'Overweight_Level_I':2,
                'Overweight_Level_II':3,
                'Obesity_Type_I':4,
                'Obesity_Type_II':5,
                'Obesity_Type_III':6}

target = target.map(dict_target)
target.head()
```

```python
target.value_counts() 
```

```python
combine[0], combine[1] = combine[0].astype(float), combine[1].astype(float) 
```

```python
scaler = MinMaxScaler()

for dataset in combine:
    dataset_numeric = dataset.apply(pd.to_numeric)
    dataset_scaled = scaler.fit_transform(dataset_numeric)
    dataset_scaled = pd.DataFrame(dataset_scaled, columns=dataset.columns)
    dataset_scaled = dataset_scaled.astype(float)
    dataset[:] = dataset_scaled
```

**Train The Model**

```python
X_train, X_val, y_train, y_val = train_test_split(train_df, target, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
def evaluate(clf):
    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_val)
    f1_train = f1_score(y_train, y_pred_train, average='weighted')
    f1_val = f1_score(y_val, y_pred_test, average='weighted')
    recall_train = recall_score(y_train, y_pred_train, average='weighted')
    recall_val = recall_score(y_val, y_pred_test, average='weighted')
    train_accuracy = accuracy_score(y_train, y_pred_train)
    val_accuracy = accuracy_score(y_val, y_pred_test)
    precision_train = precision_score(y_train, y_pred_train, average='weighted')
    precision_val = precision_score(y_val, y_pred_test, average='weighted')
    
    print("F1 score on train set:", round(f1_train, 4))
    print("F1 score on validation set:", round(f1_val, 4))
    print("Recall score on train set:", round(recall_train, 4))
    print("Recall score on validation set:", round(recall_val, 4))
    print("Train accuracy score:", round(train_accuracy, 4))
    print("Test accuracy score:", round(val_accuracy, 4))
    print("Precision score on train set:", round(precision_train, 4))
    print("Precision score on validation set:", round(precision_val, 4))
```

```python
clf1 = XGBClassifier()
clf1.fit(X_train, y_train)

print("XGBoost")
evaluate(clf1)
```

```python
clf4 = RandomForestClassifier()
clf4.fit(X_train, y_train)

print("Random Forest")
evaluate(clf4)
```

```python
clf6 = DecisionTreeClassifier()
clf6.fit(X_train, y_train)

print("Decision Tree")
evaluate(clf6)
```

```python
xgb_clf = XGBClassifier()
random_forest_clf = RandomForestClassifier()
Decition_tree_clf = DecisionTreeClassifier() 


estimators=[('xgb', xgb_clf),
            ('rf', random_forest_clf),
            ('Dt', Decition_tree_clf)
]

stacking_clf = StackingClassifier(
    estimators=estimators, final_estimator=LogisticRegression(solver='lbfgs', max_iter=10)
)

stacking_clf.fit(X_train, y_train)
```

```python
result_predict = pd.Series(stacking_clf.predict(test_df), name="NObeyesdad")
reverse_dict_target = {v: k for k, v in dict_target.items()}
target_decoded = result_predict.map(reverse_dict_target)
```

```python
result = pd.concat([test_id, target_decoded], axis=1)
result.to_csv("/kaggle/working/Multi_class_Prediction.csv", index=False) 
```
